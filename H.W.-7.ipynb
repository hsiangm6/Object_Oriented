```python
# Import related libraries and submodule
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from sklearn.datasets import fetch_openml
```

```python
# 1b. create a pivot table using index='num-of-doors', column='fuel-type', and values as price
data.pivot_table('price', index='num-of-doors', columns='fuel-type')
```

```python
# Q1c. Round off the values of question 1b output to two decimal places(將問題 1b 輸出的值四捨五入到小數點後兩位)

np.round(data.pivot_table('price', index='num-of-doors', columns='fuel-type'), 2)
```

```python
# Q1d. 
'''
Create a pivot table using multi-level indexing index=['num-of-doors','body-style'], 
and multi-level columns column=['fuel-type' , 'fuel-system'], and values = 'price'
'''
```

```python
data.pivot_table('price', index=['num-of-doors','body-style'], columns=['fuel-type' , 'fuel-system'])
```

```python
# 1e. fill NaN values by zero in the output of question 1d.
# hint use fill_value as the parameter

data.pivot_table('price', index=['num-of-doors','body-style'], columns=['fuel-type' , 'fuel-system'], fill_value=0)
```

```python
# 2a. create a pivot table using index = 'body-style' , columns = 'num-of-doors', values = 'price'.
#calculate mean and median using aggfunc argument
data.pivot_table('price', index='body-style', columns='num-of-doors', aggfunc=['mean', 'median'])
```

```python
# 2b. fill NaN values by zero in the output of question 2a
data.pivot_table('price', index='body-style', columns='num-of-doors', aggfunc=['mean', 'median'], fill_value=0)

```

```python
# 2c. create a pivot table using values=['price', 'horsepower'], and index= 'make'
data.pivot_table(values=['price', 'horsepower'], index='make')
```

```python
# 2d. fill NaN values by zero in the output of question 2c and round off the result to two decimal places 
np.round(data.pivot_table(values=['price', 'horsepower'], index='make', fill_value=0),2)

```

```python
# 2 e. ordered the car make by price from high to low value
data_2e=data.pivot_table(values='price', index='make')
data_2e=data_2e.reindex(data_2e['price'].sort_values(ascending=False).index)
data_2e
```
